```
import random

def generate_bits(n):
    """Generates random binary values representing Alice's secret message."""
    return [random.randint(0, 1) for i in range(n)]

def generate_bases(n):
    """Randomly selects quantum bases: '+' (Rectilinear) or 'x' (Diagonal)."""
    return [random.choice(['+', 'x']) for i in range(n)]

n = 8
alice_bits = generate_bits(n)
alice_bases = generate_bases(n)
bob_bases = generate_bases(n)

print("Alice Bits :", alice_bits)
print("Alice Bases:", alice_bases)
print("Bob Bases :", bob_bases)

shared_key = [alice_bits[i] for i in range(n) if alice_bases[i] == bob_bases[i]]
print("\nShared Secret Key:", shared_key)
```

```
import random

original_key = [random.randint(0, 1) for i in range(8)]
modified_key = []

for bit in original_key:

    if random.random() < 0.30:
        modified_key.append(1 - bit)
    else:
        modified_key.append(bit)

errors = sum(1 for i in range(len(original_key)) if original_key[i] != modified_key[i])

print("Original Key:", original_key)
print("Modified Key:", modified_key)
print(f"Errors      : {errors}")

if errors > 0:
    print("Attack Detected!")
else:
    print("Communication Secure")
```

```
alice_key = [1, 0, 1, 1, 0, 1, 0, 1]
bob_key   = [1, 0, 0, 1, 0, 1, 1, 1]
errors = 0

for i in range(len(alice_key)):
    if alice_key[i] != bob_key[i]:
        errors += 1

error_rate = (errors / len(alice_key)) * 100

print("Alice Key :", alice_key)
print("Bob Key   :", bob_key)
print("\nErrors    :", errors)
print("Error Rate:", error_rate, "%")

if error_rate > 10:
    print("Warning: Eavesdropper Detected!")
else:
    print("Secure Communication")
```
